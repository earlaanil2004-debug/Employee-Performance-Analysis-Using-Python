```python
#it is Header Cleaning
print(emp.columns.tolist())
emp.columns=emp.columns.str.strip().str.lower().str.replace("_", "")
print("Fixed Columns")
print(emp.columns.tolist())
```

```python
import matplotlib.pyplot as plt

# Create a simple scatter plot using your emp DataFrame for demonstration
plt.figure(figsize=(7, 5))
plt.scatter(emp['age'], emp['salary'])
plt.title('Age vs Salary with Custom Border')
plt.xlabel('Age')
plt.ylabel('Salary')

# Get the current Axes instance to customize the border
ax = plt.gca()

# Ensure all four spines (borders) are visible and set their properties
border_color = 'darkblue' # You can change this color
border_linewidth = 2 # You can change the line width

for spine in ax.spines.values():
    spine.set_visible(True) # Ensure visibility
    spine.set_color(border_color)
    spine.set_linewidth(border_linewidth)

plt.show()
```

```python
#1: Salary distribution by Department
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.figure(figsize=(12,5))
sns.barplot(x='department',y='salary', data=emp,estimator=np.mean,orient='vertical')
plt.title("Salary distribution by Department")
plt.xlabel("Departmnet")
plt.ylabel("Salary")
plt.show()
```

```python
#Shows a general upward trend, though there is significant variance within the same age groups depending on the department.
plt.figure(figsize=(10,6))
sns.scatterplot(x='age',y='salary',hue='department', data=emp)
plt.title("Avg Vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#A pie chart confirming that the majority of the workforce "Meets Expectations," while a small fraction is "Unsatisfactory."
plt.figure(figsize=(10,6))
plt.pie(emp['performancecategory'].value_counts(), labels=emp['performancecategory'].unique(), autopct='%1.1f%%')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#salary Distribution of Employees
import numpy as np
plt.figure(figsize=(10,5))
plt.hist(emp['salary'], bins=20, edgecolor='k')
plt.title("Salary Distribution")
plt.xlabel("SALARY")
plt.ylabel("Frequency")
plt.show()
```

```python
 #5: count of Satisfaction Level vs Performance Category
plt.figure(figsize=(10,6))
sns.countplot(x='performancecategory', hue='satisfactionlevel', data=emp)
plt.xlabel("Performance Category")
plt.ylabel("Count")
plt.title("Satisfaction Level vs Performance Category")
plt.legend(title='satisfactionlevel' ,bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#A horizontal bar chart highlighting which departments are the highest and lowest paying.
plt.figure(figsize=(10,6))
sns.barplot(x='salary', y='department', data=emp, orient='h',hue='department' ,palette='viridis')
plt.xlabel("Salary")
plt.ylabel("Department")
plt.show()
```
